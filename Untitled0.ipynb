```
%matplotlib inline
import matplotlib.pyplot as plt

#preparing data
x = [1,2,3,4,5]
y = [11,22,33,44,55]

fig, ax = plt.subplots(figsize=(10,10))
ax.plot(x,y)
ax.set(title='plot,',
        xlabel = 'x',
       ylabel='y')
```
